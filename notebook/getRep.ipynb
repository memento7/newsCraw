```python
frame = pd.read_csv('../data/test.csv')
```

```python
from konlpy.tag import Twitter
tagger = Twitter()

pos_tags = ['Noun']
stoptags = ['KoreanParticle', 'PreEomi', 'Punctuation', 'Eomi', 'Number', 'Foreign', 'URL']
def stemize_pos(text):
    return [ word for word, tag in tagger.pos(text) if tag in pos_tags]
```

```python
clusters = [17, 19, 20, 21, 22, 23, 24, 25, 26, 28]
```

```python
tokens = [ stemize_pos(frame.iloc[idx].content) for idx in clusters ]
```

```python
from collections import defaultdict
```

```python
token_dic = defaultdict(int)
token_dics = [ defaultdict(int) for _ in clusters ]
sentence = []
```

```python
for i, idx in enumerate(clusters):
    tokens = stemize_pos(frame.iloc[idx].content)
    for token in tokens:
        token_dic[token] += 1
        token_dics[i][token] += 1
        sentence.append(tokens)
```

```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
```

```python
model = Word2Vec()
```

```python
for epoch in range(1):
    model.train(sentence)
    if not epoch % 100: print (epoch, model.alpha)
```
